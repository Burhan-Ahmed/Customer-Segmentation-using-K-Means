 **Initialization**

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('/content/Mall_Customers.csv')
print(dataset)
```

```python
income_spend=dataset.iloc[:,[3,4]].values
kmean=KMeans(n_clusters=2,n_init='auto', random_state=42)
#Building the model using fit function
kmean.fit(income_spend)
build_mod=kmean.labels_
print(build_mod)
```

```python
#Both methods return the same output but this is direct
print(kmean.fit_predict(income_spend))
#Centroid for both the clusters
print(kmean.cluster_centers_)
```

```python
plt.figure(figsize=(8,8))
plt.scatter(income_spend[build_mod==0,0],income_spend[build_mod==0,1],s=50, c='red',label='Cluster 1')
plt.scatter(income_spend[build_mod==1,0],income_spend[build_mod==1,1], s=50,c='blue',label='Cluster 2')

plt.scatter(kmean.cluster_centers_[:, 0], kmean.cluster_centers_[:, 1], c='green', label='Centroid')

plt.title('KMeans Clustering')
plt.xlabel('Income')
plt.ylabel('Spend')
plt.legend()
plt.show()
```
